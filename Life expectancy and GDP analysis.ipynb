```python
# start by importing the relevant python libraries
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
# start by loading the data and viewing it to ensure it has loaded
df=pd.read_csv("all_data.csv")
print(df.head())
```

```python
# data has been loaded, time to start answering questions
#Question 1: Average life expentacy by country
ax=plt.subplot()
sns.barplot(data=df,x="Country",y="Life expectancy at birth (years)")
ax.set_title("Average Life Expenctancy by Country")
plt.axis([-0.5,6,40,90])
ax.set_xticklabels(ax.get_xticklabels(),rotation= 45, fontsize=8)
sns.set_context("notebook",font_scale=1)
plt.show()
print("From 2000-2015 the average life expectancy by country can be seen in the graph above. The country with the average highest life expectancy is Germany, followed closely by Chile. The country with the lowest average life expectancy is Zimbabwe.")
```

```python
#Question 2 Average Life expentacy by year by country
plt.figure(figsize=(14,10))

sns.barplot(data=df,x="Country",y="Life expectancy at birth (years)",hue="Year")
plt.title("Average Life Expenctancy by Year and Country")
plt.axis([-0.5,6,40,85])
sns.set_context("notebook",font_scale=1)
sns.color_palette("Set2")
sns.set_palette("deep")
plt.show()
print("Looking at average life expectancy over time for each country the trend is that average life expectancy has increased with time. Over the time period Zimbabwe has shown the largest increase in average life expectancy, starting in 2000 with an average life expectancy of 46 and in 2015 having an average life expectancy of 61, a 33% increase! ")
```

```python
# Distribution of GDP by Country
plt.figure(figsize=(8,12))
plt.subplot(2,3,1)
plt.subplots_adjust(top=1,wspace=1.2,hspace=0.2)
plt.boxplot(df.GDP[df.Country == "Chile"])
plt.title("GDP of Chile 2000-2015",y=1.05)
plt.ylabel("GDP")
plt.subplot(2,3,2)
plt.boxplot(df.GDP[df.Country == "China"])
plt.title("GDP of China 2000-2015 ",y=1.05)
plt.ylabel("GDP")
plt.subplot(2,3,3)
plt.boxplot(df.GDP[df.Country == "Germany"])
plt.title("GDP of Germany 2000-2015",y=1.05)
plt.ylabel("GDP")
plt.subplot(2,3,4)
plt.boxplot(df.GDP[df.Country == "Mexico"])
plt.title("GDP of Mexico 2000-2015",y=1.05)
plt.ylabel("GDP")
plt.subplot(2,3,5)
plt.boxplot(df.GDP[df.Country == "United States of America"])
plt.title("GDP of USA 2000-2015",y=1.05)
plt.ylabel("GDP")
plt.subplot(2,3,6)
plt.boxplot(df.GDP[df.Country == "Zimbabwe"])
plt.title("GDP of Zimbabwe 2000-2015",y=1.05)
plt.ylabel("GDP")
print("These box plots are showing the distribution of GDP for each country between 2000-2015. Unsurpisngly USA has the largest average GDP of each country at circa $14,500,000,000,000, closely followed by China. Zimbabwe has the lowest average GDP over the time period with a GDP of circa $6,500,000,000.")
```

```python
#Question 5 GDP of USA vs China 2000-2015, is there any overlap?
plt.hist(df.GDP[df.Country == "United States of America"],alpha=0.5,bins=10)
plt.hist(df.GDP[df.Country == "China"],alpha=0.5,bins=10)
plt.xlabel("GDP")
plt.show()
print("There is an overlap of GDP between USA and China. The maximum GDP of China overlaps with the miniumum GDP of USA between 2000-2015. From this graph it is unclear what year these two countries had their maximum and minumum GDP.")
```

```python
#Question 6 is GDP linked to life expectancy? 
f,ax=plt.subplots(figsize=(8,12))
sns.scatterplot(data=df,x="Life expectancy at birth (years)",y="GDP")
plt.ylim(1e10,1e13)
plt.xlim(65,85)
print("Looking at the graph there is a positve correlation between GDP and life expectancy. This is to expected as a country becomes more developed.")
```

```python
#Question 6: When was GDP at its highest and lowest by country?
# Create data frame for each country to find year when GDP is greatest
chile_df= df[df["Country"] == "Chile"]
china_df=df[df["Country"] == "China"]
germany_df=df[df["Country"] == "Germany"]
mexico_df= df[df["Country"]== "Mexico"]
USA_df= df[df["Country"] == "United States of America"]
zimbabwe_df= df[df["Country"] == "Zimbabwe"]
```
